```python
df = pd.read_csv("trainPre.csv") 
```

```python
df.head()
```

```python
df.sort_values(by='UF')
```

```python
d = { 'UF' : df['UF'], 'Ind Atividade Ano' : df['Ind Atividade Ano']}
dfuf = pd.DataFrame(data=d)
```

```python
dfuf
```

```python
dfufgroup = dfuf.groupby('UF')
```

```python
dfsum = dfufgroup.sum()
```

```python
lenlist = len(dfufgroup['UF'].value_counts().tolist())

listuf = dfufgroup['UF'].value_counts().keys().tolist()
listufs = list()
listsum = list()

l = 0
while l < lenlist:
    listufs.append(listuf[l][0])
    listsum.append(dfsum.get_values().tolist()[l][0])
    l = l + 1
    
```

```python
d = { 'UF' : listufs, 'Sum Ind Atividade' : listsum, 'Count' : dfufgroup['UF'].value_counts().tolist()}

dfg = pd.DataFrame(data=d) 
```
